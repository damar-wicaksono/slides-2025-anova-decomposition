```python
import numpy as np
import matplotlib.pyplot as plt
import uqtestfuns as uqtf
```

```python
fun = uqtf.Portfolio3D(parameters_id="Saltelli2004-3")
```

```python
xx = fun.prob_input.get_sample(10000)
xx_1 = xx.copy()
xx_1[:, 0] = 0
xx_2 = xx.copy()
xx_2[:, 2] = 0
```

```python
yy = fun(xx)
yy_1 = fun(xx_1)
yy_2 = fun(xx_2)
```

```python
fig, axs = plt.subplots(2, 1, figsize=(4, 6))
axs[0].hist(yy, bins="auto", label="original")
axs[0].hist(yy_1, bins="auto", label=r"$P_s$ fixed")
axs[0].legend(fontsize=12)
axs[1].hist(yy, bins="auto", label="original")
axs[1].hist(yy_2, bins="auto", label=r"$P_j$ fixed")
axs[1].legend(fontsize=12)
axs[1].set_xlabel("y")
plt.tight_layout()
plt.savefig("../figures/portfolio.png", dpi=300);
```
